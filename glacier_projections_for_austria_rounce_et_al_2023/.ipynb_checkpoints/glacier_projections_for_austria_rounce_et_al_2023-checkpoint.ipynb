# Future evolution of glaciers in Austria as projected by Rounce et al., 2023

```python
import pandas as pd
import matplotlib.pyplot as plt
import xarray as xr
import numpy as  np
import seaborn as sns
```

```python
### open file with country names for RGIids, this is just from all Alps
data = pd.read_hdf('./data/rgi62_era5_itmix_country_df.h5', 'df') # file comes from Fabien Maussion
data['vol_itmix_km3'] = data['vol_itmix_m3'] * 1e-9
data = data.loc[data.O1Region == '11']
data = data.loc[(data.CenLon > 2.5) & (data.CenLat > 43.2)]
data.plot(kind='scatter', x='CenLon', y='CenLat');
```

```python
# select austrian glaciers 
rgi_ids_austria_all = data.loc[data.Country_Name=='Austria'].index
```

```python
# get the global mean temp. file list per GCM - SSP combination 
path_temp = 'data/Rounce_et_al_2023/Global_mean_temp_deviation_2081_2100_rel_1850_1900.csv'
pd_temp_ssp_gcm = pd.read_csv(path_temp)
```

```python
# read in the per-glacier mass projection data from rounce et al 2023 of the alps
ds_l = []
for ssp in ['ssp119','ssp126', 'ssp245', 'ssp370', 'ssp585']:
    _ds = xr.open_dataset(f'data/Rounce_et_al_2023/R11_glac_mass_annual_50sets_2000_2100-{ssp}.nc')
    _ds = _ds.rename_dims({'glacier':'rgi_id', 'model':'gcm'})
    _ds = _ds['glac_mass_annual']
    _ds['ssp'] = ssp
    _ds = _ds.set_index({'rgi_id':'RGIId'})
    _ds = _ds.set_index({'gcm':'Climate_Model'})
    ds_l.append(_ds)
ds = xr.concat(ds_l, dim ='ssp')
#gcms_real = ds.gcm.T[ds.sum(dim='rgi_id').sum(dim='year') != 0]
#ds = ds.sel(gcm = gcms_real)
ds_stack = ds.stack(gcm_ssp = ('gcm', 'ssp'))
ds_stack.coords['gcm_ssp'] = [a[0]+'_'+a[1] for a in ds_stack.gcm_ssp.values]
#pd_temp_ch['gcm'] = [a.lower() for a in pd_temp_ch['GCM']]
pd_temp_ssp_gcm['gcm_ssp'] = pd_temp_ssp_gcm['GCM'] +'_'+pd_temp_ssp_gcm['Scenario']
ds_stack = ds_stack.sel(gcm_ssp=pd_temp_ssp_gcm['gcm_ssp'].unique())
ds_stack['gcm_ssp'] = [x.lower() for x in ds_stack.gcm_ssp.values]

# only select the non-failing glaciers (if a glacier fails in just one GCM/SSP combination drop it)        
for r in data.index:
    if r not in ds_stack.rgi_id.values:
        print(r)
        data = data.drop(r)
rgi_ids_austria_common_running = data.loc[data.Country_Name=='Austria'].index
```

```python
# ok, but in austria there are no failing glaciers
assert (rgi_ids_austria_all == rgi_ids_austria_common_running).all()
```

```python
# select all austrian glacier mass projections and sum them up
ds_mass_austria_sum = ds_stack.sel(rgi_id=rgi_ids_austria_all).sum(dim='rgi_id')
```

```python
y0 = 2024 # can be changed ... 

plt.rc('font', size=22)

#dt_ssp_all_avg = dt_ssp_all['glac_mass_annual'].rolling(time=1, center=True).mean()
_reg_sel = ds_mass_austria_sum

### https://pyam-iamc.readthedocs.io/en/stable/tutorials/ipcc_colors.html
colors = ['#00a9cf', # '#003466', 
          '#f69320', '#df0000', '#980002'] 

for language in ['german','english']:

    hue_order = ['+1.5°C', '+2.0°C', '+3.0°C', '+4.0°C']
    for version in ['rounce']:  # sm ### <- other option how to define it , but we keep it here as defied by DR.. 
        pd_vol_reg_rounce_rel = (_reg_sel*100/_reg_sel.sel(year=y0)).to_dataframe('rel_glac_mass').copy()  # .sum(dim='rgi_id')
        pd_vol_reg_rounce_rel = pd_vol_reg_rounce_rel.reset_index()
        plt.figure(figsize=(12,6))

        for j,h in enumerate(hue_order):
            if h=='+1.5°C':
                condi1 = (pd_temp_ssp_gcm['global_mean_deviation_degC'] < 1.75) & (pd_temp_ssp_gcm['global_mean_deviation_degC'] >= 1.25)
                ssp_gcm_l_1_5 = pd_temp_ssp_gcm.where(condi1).dropna()[['Scenario', 'GCM']].values
                for ssp,gcm in ssp_gcm_l_1_5:
                    ind = pd_vol_reg_rounce_rel.loc[pd_vol_reg_rounce_rel['gcm_ssp'] == f'{gcm}_{ssp}'.lower()].index
                    pd_vol_reg_rounce_rel.loc[ind, 'temp_change_2100'] = h
                pd_vol_reg_rounce_rel_med = pd_vol_reg_rounce_rel.groupby(['year', 'temp_change_2100']).mean().reset_index()
                pd_sel = pd_vol_reg_rounce_rel_med.loc[pd_vol_reg_rounce_rel_med.temp_change_2100 == h]
            elif h == '+2.0°C':
                if version == 'sm':
                    # another way, but results in less GCMs & SSPs (around 9)
                    condi1 = (pd_temp_ssp_gcm['global_mean_deviation_degC'] < 2.33) & (pd_temp_ssp_gcm['global_mean_deviation_degC'] >= 1.66)
                else:
                    # this is how it is defined in Rounce et al., (2023)
                    condi1 = (pd_temp_ssp_gcm['global_mean_deviation_degC'] < 2.5) & (pd_temp_ssp_gcm['global_mean_deviation_degC'] >= 1.5)

                ssp_gcm_l_1_5 = pd_temp_ssp_gcm.where(condi1).dropna()[['Scenario', 'GCM']].values
                for ssp,gcm in ssp_gcm_l_1_5:
                    ind = pd_vol_reg_rounce_rel.loc[pd_vol_reg_rounce_rel['gcm_ssp'] == f'{gcm}_{ssp}'.lower()].index
                    pd_vol_reg_rounce_rel.loc[ind, 'temp_change_2100'] = h
                pd_vol_reg_rounce_rel_med = pd_vol_reg_rounce_rel.groupby(['year', 'temp_change_2100']).mean().reset_index()
                pd_sel = pd_vol_reg_rounce_rel_med.loc[pd_vol_reg_rounce_rel_med.temp_change_2100 == h]
            elif h == '+3.0°C':
                condi1 = (pd_temp_ssp_gcm['global_mean_deviation_degC'] < 3.5) & (pd_temp_ssp_gcm['global_mean_deviation_degC'] >= 2.5)
                ssp_gcm_l_1_5 = pd_temp_ssp_gcm.where(condi1).dropna()[['Scenario', 'GCM']].values
                for ssp,gcm in ssp_gcm_l_1_5:
                    ind = pd_vol_reg_rounce_rel.loc[pd_vol_reg_rounce_rel['gcm_ssp'] == f'{gcm}_{ssp}'.lower()].index
                    pd_vol_reg_rounce_rel.loc[ind, 'temp_change_2100'] = h
                pd_vol_reg_rounce_rel_med = pd_vol_reg_rounce_rel.groupby(['year', 'temp_change_2100']).mean().reset_index()
                pd_sel = pd_vol_reg_rounce_rel_med.loc[pd_vol_reg_rounce_rel_med.temp_change_2100 == h]
            elif h == '+4.0°C':
                condi1 = (pd_temp_ssp_gcm['global_mean_deviation_degC'] < 4.5) & (pd_temp_ssp_gcm['global_mean_deviation_degC'] >= 3.5)
                ssp_gcm_l_1_5 = pd_temp_ssp_gcm.where(condi1).dropna()[['Scenario', 'GCM']].values
                for ssp,gcm in ssp_gcm_l_1_5:
                    ind = pd_vol_reg_rounce_rel.loc[pd_vol_reg_rounce_rel['gcm_ssp'] == f'{gcm}_{ssp}'.lower()].index
                    pd_vol_reg_rounce_rel.loc[ind, 'temp_change_2100'] = h
                pd_vol_reg_rounce_rel_med = pd_vol_reg_rounce_rel.groupby(['year', 'temp_change_2100']).mean().reset_index()
                pd_sel = pd_vol_reg_rounce_rel_med.loc[pd_vol_reg_rounce_rel_med.temp_change_2100 == h]    
            _p = pd_vol_reg_rounce_rel.loc[pd_vol_reg_rounce_rel.temp_change_2100 == h]
            print(h, len(_p.loc[_p.year==2000]))

            sns.lineplot(x='year', y='rel_glac_mass',
                         #hue='temp_change_2100',
                         color = colors[j], #colors_ssp,
                         label=h,
                         data= pd_sel.loc[pd_sel.year>=y0],
                         lw = 5, alpha = 0.9)

        plt.axhline(0, color='grey', alpha =0.3, lw=2)
        plt.axhline(50, color='grey', alpha =0.3, lw=1)
        plt.axhline(100, color='grey', alpha =0.3, lw=2)
        plt.ylim([0,132])


        sns.lineplot(x='year', y='rel_glac_mass',
                     data=pd_vol_reg_rounce_rel_med.loc[pd_vol_reg_rounce_rel_med.year<=y0].groupby('year').median().reset_index(),
                     color = 'black', 
                     lw=5, alpha = 0.9)
        if language == 'german':
            plt.ylabel(f'Gletschermasse Österreichs\n % relativ zu {y0}') #, fontweight='bold')
            plt.xlabel('Jahr') #, fontweight='bold');
        else:
            plt.ylabel(f'Glacier mass of Austria\n % relative to {y0}') #, fontweight='bold')
            plt.xlabel('Year') #, fontweight='bold');
        plt.axvline(y0, color ='grey', ls='--', lw=2, zorder=0, alpha = 0.6)
        
        # could also start in 2000, but then a lot of the past loss is just visible... 
        #plt.xlim([1999.5,2100.5])
        plt.xlim([2015,2100])
        if language == 'german':
            leg = plt.legend(title='Globale Erwärmung\nbis 2100', loc = 'upper right')
        elif language == 'english':
            leg = plt.legend(title='Global warming above\npre-industrial by 2100', loc = 'upper right')
            
        # get the individual lines inside legend and set line width
        for line in leg.get_lines():
            line.set_linewidth(5)

        #ax2 = plt.gca().secondary_yaxis("right")

        plt.gca().tick_params(axis='y', which='major', right=False, width=2, length=5)
        plt.gca().tick_params(axis='x', which='major', bottom=True, width=2, length=5)
        plt.gca().tick_params(axis='x', which='minor', bottom=True, width=1, length=3)
        plt.gca().tick_params(axis='y', which='major', left=True, width=2, length=5)
        plt.gca().tick_params(axis='y', which='minor', right=False, width=1, length=3)
        ax2.set_yticklabels(labels=['','','','','',''])

        plt.gca().tick_params(axis='y', which='minor', left=True, width=1, length=3)
        max_2100 = int(np.round(pd_vol_reg_rounce_rel.dropna().groupby('temp_change_2100').min().rel_glac_mass.max()))
        if language == 'german':
            text = f'...weniger als {max_2100}%\nverbleibt bis 2100'
        elif language == 'english':
            text = f'...less than {max_2100}%\nremains until 2100'
            
        plt.annotate(text,
                     (2079,12),
                 color='black', fontsize=18, zorder=99)
        plt.gca().minorticks_on()
        plt.tight_layout()
        plt.gca().spines['top'].set_visible(False)
        plt.gca().spines['right'].set_visible(False)
        median_yr_half_remaining = pd_vol_reg_rounce_rel.loc[(pd_vol_reg_rounce_rel.rel_glac_mass<51) & (pd_vol_reg_rounce_rel.rel_glac_mass>49)].dropna().median().year
        years_to_lose_half = int(median_yr_half_remaining-y0)
        
        if language == 'german':
            text = f"etwa die\nHälfte\nschmilzt in\n{years_to_lose_half} Jahren"
        elif language == 'english':
            text = f"about half\nwill melt\nin the next\n{years_to_lose_half} years"
        t = plt.gca().text(
            y0+0.5, 22, text, ha="left", va="center", rotation=0, size=14,
            bbox=dict(boxstyle="rarrow,pad=0.3", fc="silver", ec="grey", lw=2, alpha = 0.5))
        # range...
        q1, q2 = pd_vol_reg_rounce_rel.loc[(pd_vol_reg_rounce_rel.rel_glac_mass<51) & (pd_vol_reg_rounce_rel.rel_glac_mass>49)].dropna().quantile([0.17,0.83]).year.values
        plt.axvspan(q1,q2, alpha = 0.2, zorder=99, color='silver')

        #plt.axvline(2041)
        if version == 'sm':
            plt.savefig('rounce_2023_austria_german_figure_median_2015_1_66_2_33.png', bbox_inches='tight', dpi=300)
        else:
            plt.savefig(f'rounce_2023_austria_{language}_figure_median_rounce_def.png', bbox_inches='tight', dpi=300)
            plt.savefig(f'rounce_2023_austria_{language}_figure_median_rounce_def.pdf', bbox_inches='tight', dpi=300)

        #plt.savefig('rounce_2023_alps_german_figure_mean_2015_rounce_def.pdf')
```
